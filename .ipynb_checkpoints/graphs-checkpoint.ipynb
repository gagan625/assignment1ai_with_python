# Drawing Graphs

The above about drawing also applies to drawing graphs of mathematical functions. 

```python
x = np.linspace(0,7,100)
y = np.sin(x)
plt.plot(x,y, label='sin')
#plt.show()

# Adding titles and captions
plt.title('Title')
plt.xlabel('x')
plt.ylabel('y')

# Changing the color
#plt.plot(x,y,"r") 
# plt.plot(x,y,color="#4b0082")

# Changing the marker
#plt.plot(x,y,"o")

# Changing the linestyle and the linewidth
#plt.plot(x,y,linewidth=2,linestyle="--")

#To get more than one graph on the same image, use the plot () command to enter more points
plt.plot(x,y,"go",x,2*y,"b^",label='cos')

# To explain multiple graphs, used legend method
plt.legend(loc="lower right")
#plt.legend(['sin(x)','cos(x)'])



plt.show()
```

Another way to display multiple graphs in the same figure window is to divide the window into subplots.

```python
plt.subplot(1,2,1) # 1x2 grid 1. subpicture
plt.plot(x,y)
plt.title("first")
plt.subplot(1,2,2) # 1x2 grid 2. subpicture
plt.plot(x,2*y)
plt.title("second")
plt.suptitle("Common Title")
plt.show()
```

#### Histograms

In data sciences and machine learning, statistical distributions of data are often studied. These can be quickly illustrated with histograms.

```python
x = np.random.randn(2000)
plt.hist(x,10)
plt.ylabel('frequencies')
plt.show()
```

#### Scatter plots

Scattering pattern refers only to the representation of individual data points with dots without lines. 

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.scatter(x,y)
#plt.scatter(x,y,color="r",marker="o",label="Points") # additional attributes can be used 
#to adjust the colors of the scattering pattern
plt.show()
```

#### Multidimestional graphs

Graphs can also be graphs of functions that depend on two variables. 

```python
points = np.arange(-2,2,0.01)
x,y = np.meshgrid(points,points)
z = np.sqrt(x**2 + y**2)
plt.imshow(z)
plt.colorbar()
plt.show()
```

As a side note, imshow () can also be used to display the contents of image files as long as the image file is first read in numeric format. e.g.,

```python
import matplotlib.image as mpimg
I=mpimg.imread('chap2-cameraman.png')
plt.imshow(I)
```

```python
#Drawing three-dimensional textures also requires the mpl_toolkits module. 
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.linspace(-2, 2, 30),np.linspace(-2, 2, 30))
z = np.cos(x ** 2 + y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z)
ax.set_title('Texture')
# plt.savefig('picture.pdf')
plt.show()
```
